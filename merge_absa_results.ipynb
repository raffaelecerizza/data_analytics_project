```python
import numpy as np
import pandas as pd
```

# Constants

```python
PREPROCESSED_DATASET_PATH = "./dataset/dataset.csv"
ABSA_REVIEWS_PATH = "./dataset/absa_reviews.csv"
ABSA_DATASET_PATH = "./dataset/absa_dataset.csv"
DATASET_WITH_SENTIMENT_PATH = "./dataset/dataset_with_sentiment.csv"
ABSA_RESULT_1 = "./absa_results/absa_proc_dataset [0-100000].csv"
ABSA_RESULT_2 = "./absa_results/absa_proc_dataset [100000-200000].csv"
ABSA_RESULT_3 = "./absa_results/absa_proc_dataset [200000-300000].csv"
ABSA_RESULT_4 = "./absa_results/absa_proc_dataset [300000-400000].csv"
ABSA_RESULT_5 = "./absa_results/absa_proc_dataset [400000-500000].csv"
ABSA_RESULT_6 = "./absa_results/absa_proc_dataset [500000-600000].csv"
ABSA_RESULT_7 = "./absa_results/absa_proc_dataset [600000-700000].csv"
ABSA_RESULT_8 = "./absa_results/absa_proc_dataset [700000-end].csv"
```

# Import and merge ABSA results

```python
df_absa_result_1 = pd.read_csv(ABSA_RESULT_1)
df_absa_result_2 = pd.read_csv(ABSA_RESULT_2)
df_absa_result_3 = pd.read_csv(ABSA_RESULT_3)
df_absa_result_4 = pd.read_csv(ABSA_RESULT_4)
df_absa_result_5 = pd.read_csv(ABSA_RESULT_5)
df_absa_result_6 = pd.read_csv(ABSA_RESULT_6)
df_absa_result_7 = pd.read_csv(ABSA_RESULT_7)
df_absa_result_8 = pd.read_csv(ABSA_RESULT_8)
```

```python
absa_dataset = pd.concat([df_absa_result_1, df_absa_result_2, df_absa_result_3,
                          df_absa_result_4, df_absa_result_5, df_absa_result_6,
                          df_absa_result_7, df_absa_result_8])
```

```python
absa_dataset.to_csv(ABSA_DATASET_PATH, encoding='utf-8', index=False)
```
